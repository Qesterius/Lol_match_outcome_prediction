```python
import pandas as pd
```

```python
ts_df = pd.read_csv("./data_csv/timestamps.csv")
ts_df.shape
```

```python
ts_df.head(20)
```

```python
m_df = pd.read_csv("./data_csv/matches.csv")
m_df.shape
```

```python
selected_cols = ['matchId', 'gameDuration'] + [f"p{i}_championId" for i in range(10)]

selected_cols
```

```python
m_df_selection = m_df[selected_cols] 
m_df_selection.head()
```

```python
m_df_selection.shape
```

```python
ts_df_sample = ts_df.head(50)
ts_df_sample.head()
```

```python
df_merged = pd.merge(ts_df_sample, m_df_selection, on='matchId', how='inner')
```

```python
df_merged.head(20)
```

```python
for i in range(10, 0, -1):
    df_merged.rename(columns={f"p{i-1}_championId": f"p{i}_championId"}, inplace=True)
```

```python
df_merged.head(20)
```

```python
df_merged['gameDuration']/60
```

```python
df_merged['percentTimeStamp'] = df_merged['timeStamp']*60/df_merged['gameDuration']
df_merged[['percentTimeStamp', 'gameDuration', 'timeStamp']][df_merged['percentTimeStamp'] >= 1]
```

```python
df_merged
```

```python
df_merged_all = pd.merge(ts_df, m_df_selection, on='matchId', how='inner')

for i in range(10, 0, -1):
    df_merged_all.rename(columns={f"p{i-1}_championId": f"p{i}_championId"}, inplace=True)
    
df_merged_all['percentTimeStamp'] = df_merged_all['timeStamp']*60/df_merged_all['gameDuration']
```

```python
df_merged_all
```

```python
df_merged_all.to_csv("./data_csv/timestamps_ext.csv", index=False)
```
